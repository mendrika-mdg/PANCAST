```python
from datetime import datetime, timedelta

start = datetime(2025, 7, 1, 0, 0)
end = datetime(2025, 7, 14, 23, 59)

time_list = []
dt = start
while dt <= end:
    time_list.append({
        "year": f"{dt.year}",
        "month": f"{dt.month:02d}",
        "day": f"{dt.day:02d}",
        "hour": f"{dt.hour:02d}",
        "minute": f"{dt.minute:02d}",
    })
    dt += timedelta(minutes=15)

# Optional: check length
print(f"Total timestamps: {len(time_list)}")
```
